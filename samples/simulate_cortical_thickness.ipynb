# How to simulate 3D data for the cortical surface

## you could report this as a z-score (set bounds for simulations -1:1)

```python
subject = 'S1'
xfm = 'fullhead'

# for example, get vertices for fusiform face area FFA in subject S1
roi_verts = cortex.get_roi_verts('S1', 'FFA')

# roi_verts is a dictionary (in this case with only one entry)
ffa_verts = roi_verts['FFA']

# this includes indices from both hemispheres
# let's create a Vertex object and fill FFA
ffa_map = cortex.Vertex.empty('S1', cmap='coolwarm')
ffa_map.data = np.random.randn(ffa_map.data.shape[0], ) # default random noise 
ffa_map.data[ffa_verts] = ffa_map.data[ffa_verts]+1 # change this as needed for your predicted results 
# ffa_map.data[ffa_verts] = ffa_map.data[ffa_verts]+1000 # dramatic


ffa_map.data[ffa_map.data < 0.8] = np.nan # thresholding to get rid of the tiny values if you need it

# This creates a Volume object for our test dataset for the given subject
# and transform
cortex.quickshow(ffa_map, cvmin=0.8, cvmax=1., curvature_threshold=True)
plt.show()
```
